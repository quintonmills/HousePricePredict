###Machine Learning Pipeline - Feature Selection

In this notebook we pick up transformed datasets that we saved in the previos notebook

###Reproducibility: Setting the seed

With the aim to ensure repreoducibility between runs of the same notebook, but also between the research and production enviornment, for each step that includes some element of randomness, it is important to set the seed

```python
#To handle datasets
import pandas as pd
import numpy as np

#for plotting
import matplotlib.pyplot as plt

#to build the models
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel

#to visualise all the columns in the dataframe
pd.pandas.set_option('display.max_columns', None)
```

```python
#Load train and test set with the engineered variables

#We built and saved these datasets in the previous notebook.

x_train = pd.read_csv('xtrain.csv')
x_test = pd.read_csv('xtest.csv')

x_train.head()
```

```python
#load the target (the target is log transformed)

y_train = pd.read_csv('ytrain.csv')
y_test = pd.read_csv('ytest.csv')
y_train.head()
```

###Feature Selection

Lets go ahead and select a subset of the most predictive features. There is an element of randomness in the Lasso regression

```python
#We will do the model fitting and feature selection altogether in a few lines of code

#First we specify the Lasso Regression Model, and we select a suitable alpha
#(equivalent of penalty). The bigger the alpha the less features that will be selected

#then we use the selectFromModel object from sklearn, which will select automatically
#the features which coefficients are non-zero

sel_ = SelectFromModel(Lasso(alpha = 0.001, random_state = 0))

#Train Lasso model and select features
sel_.fit(x_train, y_train)
```

```python
sel_.get_support().sum()
```

```python
#Lets visualise those features that were selected
#(selected featues marked with True)
sel_.get_support()
```

```python
#Lets print the total number of total and selected features

#This is how we can make a list of the selected featuers
selected_feats = x_train.columns[(sel_.get_support())]

#Lets print some stats
print('total features: {}'.format((x_train.shape[1])))
print('selected features: {}'.format(len(selected_feats)))
print('features with coefficients shrank to zero: {}'.format(
    np.sum(sel_.estimator_.coef_ == 0)))
```

```python
#Print selected features
selected_feats
```

```python
pd.Series(selected_feats).to_csv('selected_features.csv', index = False)
```
